```python
from sklearn.neighbors import KNeighborsRegressor as knn
from sklearn.model_selection import train_test_split

train, test = train_test_split(df2,
                               test_size=0.2, 
                               random_state=42,
                               shuffle=True)
print("The size of the housing dataset", df.shape[0])
print("The size of the train dataset", train.shape[0])
print("The size of the test dataset",test.shape[0] )

def test_and_train(columnName: str):
    

    columnScore = columnName + "score"
    scoretrain=[knn(n_neighbors=i).fit(train[[columnName]],
                train['ModelYear']).score(train[[columnName]],
                train['ModelYear']) for i in list(range(5, 300))]

    trainscore=pd.DataFrame(scoretrain, columns=[columnScore])
    
    trainscore['datsplit']='Training'
    trainscore['neighbors']=list(range(5, 300))

    scoretest=[knn(n_neighbors=i).fit(train[[columnName]],
                train['ModelYear']).score(test[[columnName]],
                test['ModelYear']) for i in list(range(5, 300))]

    testscore=pd.DataFrame(scoretest, columns=[columnScore])
    
    testscore['datsplit']='Testing'
    testscore['neighbors']=list(range(5, 300))

    score_df = pd.concat([trainscore, testscore], axis=0)

    (ggplot(score_df, aes(x='neighbors', y=columnScore, color='datsplit'))+
    geom_line()+
    scale_color_manual(values=['darkorange', 'steelblue'])+
     theme_bw()
    )

    print(testscore[columnScore].max())
    print(testscore.loc[testscore[columnScore]==testscore[columnScore].max(), 
                  'neighbors'])
```

```python
for col in df2.columns:
    # only run test/fit on variable columns, not prediction
    if(col != "ModelYear"):
        print(col)
        print("Score:")
        test_and_train(col)
        print()
    
    
```

# Location Predictor (North to South)

```python
df2
# What regions are more likely to have a car stolen if it's from a given model year?

# 'Canterbury', 'Wellington', 'Auckland', 'Bay of Plenty','Northland', 'Waikato', "Hawke's Bay", 'Southland', 'Tasman'

# Assign a location index value for each region (moving north to south)
conditions = [
    (df['Location'] == "Northland"),
    (df['Location'] == "Auckland"),
    (df['Location'] == "Bay of Plenty"),
    (df['Location'] == "Hawke's Bay"),
    (df['Location'] == "Waikato"),
    (df['Location'] == "Wellington"),
    (df['Location'] == "Tasman"),
    (df["Location"] == "Canterbury"),
    (df["Location"] == "Southland")
    ]

# create a list of the values we want to assign for each condition
values = [1,2,3,4,5,6,7,8,9]

# create a new column and use np.select to assign values to it using our lists as arguments
df2['LocationIndex'] = np.select(conditions, values)

df2
```

```python

msk = np.random.rand(len(df)) < 0.8

train = df2[msk]
test = df2[~msk]
print(len(train), len(test))


df2["DateStolen"].dtype
```

```python

columnName = "DateStolen"
columnScore = columnName + " score"


scoretrain=[knn(n_neighbors=i).fit(train[[columnName]],
            train['LocationIndex']).score(train[[columnName]],
            train['LocationIndex']) for i in list(range(5, 200))]

trainscore=pd.DataFrame(scoretrain, columns=[columnScore])

trainscore['datsplit']='Training'
trainscore['neighbors']=list(range(5, 200))

scoretest=[knn(n_neighbors=i).fit(train[[columnName]],
            train['LocationIndex']).score(test[[columnName]],
            test['LocationIndex']) for i in list(range(5, 200))]

testscore=pd.DataFrame(scoretest, columns=[columnScore])

testscore['datsplit']='Testing'
testscore['neighbors']=list(range(5, 200))

score_df = pd.concat([trainscore, testscore], axis=0)

print(testscore[columnScore].max())
print(testscore.loc[testscore[columnScore]==testscore[columnScore].max(), 
                  'neighbors'])

(ggplot(score_df, aes(x='neighbors', y=columnScore, color='datsplit'))+
geom_line()+
scale_color_manual(values=['darkorange', 'steelblue'])+
 theme_bw()
)

```

```python
df["VehicleModel"].unique()

d = {}

i = 0
for name in df["VehicleModel"].unique():
    d[name] = i
    i+=1


df['MakeIndex'] = df['VehicleModel'].map(d)
df
    
```

```python
df["Color"].unique()

d = {}

i = 0
for name in df["Color"].unique():
    d[name] = i
    i+=1


df['ColorIndex'] = df['Color'].map(d)
df
    
```

```python
df["Location"].unique()

d = {}

i = 0
for name in df["Location"].unique():
    d[name] = i
    i+=1


df['RegionIndex'] = df['Location'].map(d)
df
    
    
df["VehicleType"].unique()

d = {}

i = 0
for name in df["VehicleType"].unique():
    d[name] = i
    i+=1


df['TypeIndex'] = df['VehicleType'].map(d)
df
    
```

```python


X = df.drop("VehicleModel", axis=1).drop("Color", axis=1).drop("VehicleDesc", axis=1).drop("VehicleType", axis=1).drop("Location", axis=1).drop("Region_x", axis=1).drop("Region_y", axis=1).drop("DateStolen", axis=1).drop("YearHalf", axis=1)
X
y = df["YearHalf"]

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier

feature_names = [f"feature {i}" for i in range(X.shape[1])]
forest = RandomForestClassifier(random_state=0)
forest.fit(X_train, y_train)
```

```python
import time
import numpy as np

start_time = time.time()
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
elapsed_time = time.time() - start_time

print(f"Elapsed time to compute the importances: {elapsed_time:.3f} seconds")
```

```python
import pandas as pd

forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig
```
